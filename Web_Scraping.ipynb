# Web Scraping with Beautiful Soup

```python
# Import the necessary libraries
import requests
from bs4 import BeautifulSoup

# The URL of the eBay page from which we want to scrape data
url = 'https://www.ebay.com/sch/i.html?_nkw=solar+panels'

# Send an HTTP GET request to the URL and get the response
response = requests.get(url)

# Parse the HTML content of the response using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the list items (li) with the class 's-item' on the page
items = soup.find_all('li', class_='s-item')

# Loop through each item in the 'items' list
for item in items:
    # Find the title element within the current item using the class 's-item__title'
    title_element = item.find('div', class_='s-item__title')
      
    # Extract the text of the title element and remove any leading or trailing whitespace
    # If the title_element is not found, set the 'title' to "N/A"
    title = title_element.text.strip() if title_element else "N/A"

    # Find the price element within the current item using the class 's-item__price'
    price_element = item.find('span', class_='s-item__price')

    # Extract the text of the price element and remove any leading or trailing whitespace
    # If the price_element is not found, set the 'price' to "N/A"
    price = price_element.text.strip() if price_element else "N/A"

    # Print the extracted title and price information for the current item
    print(f"Title: {title}")
    print(f"Price: {price}")
    print('---')

# The URL of the eBay page from which we want to scrape data
url = 'https://www.ebay.com/sch/i.html?_nkw=solar+panels'

# Send an HTTP GET request to the URL and get the response
response = requests.get(url)

# Parse the HTML content of the response using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the list items (li) with the class 's-item' on the page
items = soup.find_all('li', class_='s-item')

# Open a CSV file in write mode to store the scraped data
csv_file = open('ebay_solar_panels.csv', 'w', newline='', encoding='utf-8')

# Create a CSV writer object
```

# Web Scraping and Saving to CSV

```python
# Import the necessary libraries
import requests
from bs4 import BeautifulSoup
import csv

# The URL of the eBay page from which we want to scrape data
url = 'https://www.ebay.com/sch/i.html?_nkw=solar+panels'

# Send an HTTP GET request to the URL and get the response
response = requests.get(url)

# Parse the HTML content of the response using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the list items (li) with the class 's-item' on the page
items = soup.find_all('li', class_='s-item')

# Open a CSV file in write mode to store the scraped data
csv_file = open('C:\\Users\\prasa\\OneDrive\\Documents\\Analysis of profit & loss\\Web_Scraping\\ebay_solar_panels.csv', 'w', newline='', encoding='utf-8')

# Create a CSV writer object
csv_writer = csv.writer(csv_file)

# Write the header row to the CSV file
csv_writer.writerow(['Title', 'Price'])

# Loop through each item in the 'items' list
for item in items:
    # Find the title element within the current item using the class 's-item__title'
    title_element = item.find('div', class_='s-item__title')
    
    # Extract the text of the title element and remove any leading or trailing whitespace
    # If the title_element is not found, set the 'title' to "N/A"
    title = title_element.text.strip() if title_element else "N/A"

    # Find the price element within the current item using the class 's-item__price'
    price_element = item.find('span', class_='s-item__price')

    # Extract the text of the price element and remove any leading or trailing whitespace
    # If the price_element is not found, set the 'price' to "N/A"
    price = price_element.text.strip() if price_element else "N/A"

    # Write the title and price data for the current item to the CSV file
    csv_writer.writerow([title, price])

# Close the CSV file
csv_file.close()

# Print a message indicating that the data has been saved in the CSV file
print("Data has been saved in ebay_solar_panels.csv file.")



```
